```python
import CMGDB

import matplotlib.pyplot as plt
import numpy as np
import time
```

# Periodic domain example

```python
def f(x):
    return [((x[0] - 0.5)**3 + 0.5) % 1, (np.cbrt((x[1] - 0.5)) + 0.5) % 1]

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```

```python
subdiv_min = 15
subdiv_max = 20
lower_bounds = [0.0, 0.0]
upper_bounds = [1.0, 1.0]
phase_periodic = [True, True]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, phase_periodic, F)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph, xlim=[0, 1], ylim=[0, 1])
```

# Example 2

```python
def f(x):
    y = [((x[0] - 0.5)**3 + 0.5) % 1, np.cbrt((x[1] - 0.5)) + 0.5]
    if y[1] >= 1:
        y[1] = 1
    if y[1] <= 0:
        y[1] = 0
    return y

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```

```python
subdiv_min = 18
subdiv_max = 25
lower_bounds = [0.0, 0.0]
upper_bounds = [1.0, 1.0]
phase_periodic = [True, False]
# phase_periodic = [True, True]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, phase_periodic, F)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph)
```

# Example 6

```python
# Define the torus map
# f(x, y) = (2 x + 3 y, 0.5 x + 0.3 y) mod 1
def f(x):
    return [(2.0 * x[0] + 3.0 * x[1]) % 1, (0.5 * x[0] + 0.3 * x[1]) % 1]

# Define box map for f
def F(rect):
    return CMGDB.BoxMap(f, rect, padding=True)
```

```python
subdiv_min = 20
subdiv_max = 30
lower_bounds = [0.0, 0.0]
upper_bounds = [1.0, 1.0]
phase_periodic = [True, True]

model = CMGDB.Model(subdiv_min, subdiv_max, lower_bounds, upper_bounds, phase_periodic, F)
```

```python
%%time
morse_graph, map_graph = CMGDB.ComputeMorseGraph(model)
```

```python
CMGDB.PlotMorseGraph(morse_graph)
```

```python
CMGDB.PlotMorseSets(morse_graph)
```
